## Invarianza a la perspectiva

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def corners(img):
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    corners = cv2.goodFeaturesToTrack(gray,8,0.01,10)
    corners = np.int0(corners)
    for i in corners:
        x,y = i.ravel()
        cv2.circle(img,(x,y),10,255,-1)
    

img = cv2.imread('nave01.png')
img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
corners(img)

img2 = cv2.imread('nave02.png')
img2= cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
corners(img2)

img3 = cv2.imread('nave03.png')
img3= cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
corners(img3)

plt.subplot(131),plt.imshow(img),plt.title('Fig1')
plt.subplot(132),plt.imshow(img2),plt.title('Fig2')
plt.subplot(133),plt.imshow(img3),plt.title('Fig3')
plt.show()
```

## Invarianza a la escala

```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

def corners(img):
    gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    corners = cv2.goodFeaturesToTrack(gray,20,0.3,30)
    corners = np.int0(corners)
    for i in corners:
        x,y = i.ravel()
        cv2.circle(img,(x,y),10,255,-1)
    
    
    
img = cv2.imread('esquina01.png')
img= cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
corners(img)

img2 = cv2.imread('esquina02.png')
img2= cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
corners(img2)

img3 = cv2.imread('esquina03.png')
img3= cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
corners(img3)

img4 = cv2.imread('esquina04.png')
img4= cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
corners(img4)

plt.subplot(221),plt.imshow(img)
plt.subplot(222),plt.imshow(img2)
plt.subplot(223),plt.imshow(img3)
plt.subplot(224),plt.imshow(img4)
plt.show()
```
