```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img1 = cv2.imread('box2.png',0)          # queryImage
img2 = cv2.imread('box_in_scene.png',0) # trainImage

#orb o sift
orb=True
```

```python
# Creamos el detector SIFT


if orb: sift = cv2.ORB_create()
else: sift = cv2.xfeatures2d.SIFT_create()
    
# obtenemos los puntos clave y los decriptores
kp1, des1 = sift.detectAndCompute(img1,None)
kp2, des2 = sift.detectAndCompute(img2,None)

print(kp1[0])
print(des1[0])
```

```python
if orb: bf = cv2.BFMatcher(normType=cv2.NORM_HAMMING) #Para descriptores ORB
else: bf = cv2.BFMatcher(normType=cv2.NORM_L2, crossCheck=False) #Para descriptores SIFT


#La k significa cuántos matches quiere hacer para cada punto, acomodados en una lista, 
#donde el mejor match está en la primera posición

matches = bf.knnMatch(des1,des2,k=2)

ejemplo=matches[2]
print(ejemplo[0].distance,ejemplo[1].distance)

punto=0
ida, idb =ejemplo[punto].queryIdx, ejemplo[punto].trainIdx

print(ida, idb)
print(kp1[ida].pt, kp1[ida].angle, kp1[ida].size, kp1[ida].octave)
print(kp2[idb].pt, kp2[idb].angle, kp2[idb].size, kp2[idb].octave)

```

```python
# Cociente de LOWE
good = []
for m,n in matches:
    if m.distance < 0.7*n.distance:
        good.append(m)
```

Encontrar transformada de perspectiva entre los puntos de la imagen original y la de prueba.

El método puede ser RANSAC o Least_Median

```python
MIN_MATCH_COUNT = 10

if len(good)>MIN_MATCH_COUNT:
    #Da el formato requerido por findHomography
    src_pts = np.float32([ kp1[m.queryIdx].pt for m in good ]).reshape(-1,1,2)
    dst_pts = np.float32([ kp2[m.trainIdx].pt for m in good ]).reshape(-1,1,2)

    M, mask = cv2.findHomography(src_pts, dst_pts, cv2.RANSAC,5.0)

    #ravel quita las dimensiones y deja un arreglo unidimensional.
    #toList convierte un arreglo a lista de valores
    matchesMask = mask.ravel().tolist()

    #obtiene las dimensiones de la primera imagen
    #reshape cambia un arreglo a un arreglo de arreglos con dimensiones (1,2) cada línea
    #pts contiene las coordenadas de las esquinas de la imagen original, al ser un rectángulo, 
    #  las esquinas simplemente son las dimensiones de la imagen.
    
    
    h,w = img1.shape
    pts = np.float32([ [0,0],[0,h-1],[w-1,h-1],[w-1,0] ]).reshape(-1,1,2)
    
    # calculamos cómo quedarían los puntos de nuestro rectángulo utilizando nuestra homografía
    dst = cv2.perspectiveTransform(pts,M)

    # dibujamos el contorno sobre la imagen de prueba utilizando un polyline
    
    img2 = cv2.polylines(img2,[np.int32(dst)],True,255,3, cv2.LINE_AA)
else:
    print ("No hubo suficientes matches - {0}/{1}".format(len(good),MIN_MATCH_COUNT))
    matchesMask = None

print(M)

```

Dibujamos las conexiones que sí hicieron match

```python
draw_params = dict(matchColor = (0,255,0), # verde
                   singlePointColor = (255,0,0),
                   matchesMask = matchesMask, # enmascarar los inliners
                   flags = 2)

img3 = cv2.drawMatches(img1,kp1,img2,kp2,good,None,**draw_params)

plt.imshow(img3, 'gray'),plt.show()
```
