# Detección de Bordes
Para detectar los bordes de la imagen, usaremos el algoritmo de Canny.
Este algoritmo realiza los siguientes pasos:

### 1. Eliminación de ruido
Los bordes son muy susceptibles al ruido, por lo que primero se aplicará un filtraje utilizando un filtro gaussiano de 5x5.

### 2. Cálculo del gradiente de la imagen
Se aplica un núcleo de Sobbel en horizontal y en vertical. Esto encontrará la primera derivada vertical y horizontal $G_x$ y $G_y$.
Para el cálculo del gradiente utilizaremos:
$$Gradiente=\sqrt{G_x^2 + G_y^2}$$
$$Ángulo=tan^{-1}\bigg(\frac{G_y}{G_x}\bigg)$$
### Remoción de máximos locales
Aunque la imagen cambie en cierto punto, si no construye un borde, no lo queremos, por lo tanto se escanea la imagen para remover máximos locales.
### Umbral de histéresis
Ya removidos los máximos, se analizan los resultados siguientes para ver si están conectados. Si están conectados, lo seguimos conectando para ir formando orillas. Si un punto está aislado, no se considerará orilla. para esto, se utilizan umbrales de intensidad. Un umbral max y uno min. Si un valor está arriba de max, se considera orilla segura. Si uno está abajo de min, se considera no-orilla segura.
Estos parámetros se pasan a openCV con la función _Canny_ (imagen, umbralMax, umbralMin)

```python
edges = cv2.Canny(img,180,250, apertureSize=3)
plt.imshow(edges,cmap = 'gray')
plt.show()
```
