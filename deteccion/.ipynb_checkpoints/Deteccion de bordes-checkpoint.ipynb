# Deteccin de Bordes
Para detectar los bordes de la imagen, usaremos el algoritmo de Canny.
Este algoritmo realiza los siguientes pasos:

### 1. Eliminacin de ruido
Los bordes son muy susceptibles al ruido, por lo que primero se aplicar un filtraje utilizando un filtro gaussiano de 5x5.

### 2. Clculo del gradiente de la imagen
Se aplica un ncleo de Sobbel en horizontal y en vertical. Esto encontrar la primera derivada vertical y horizontal $G_x$ y $G_y$.
Para el clculo del gradiente utilizaremos:
$$Gradiente=\sqrt{G_x^2 + G_y^2}$$
$$ngulo=tan^{-1}\bigg(\frac{G_y}{G_x}\bigg)$$
### Remocin de mximos locales
Aunque la imagen cambie en cierto punto, si no construye un borde, no lo queremos, por lo tanto se escanea la imagen para remover mximos locales.
### Umbral de histresis
Ya removidos los mximos, se analizan los resultados siguientes para ver si estn conectados. Si estn conectados, lo seguimos conectando para ir formando orillas. Si un punto est aislado, no se considerar orilla. para esto, se utilizan umbrales de intensidad. Un umbral max y uno min. Si un valor est arriba de max, se considera orilla segura. Si uno est abajo de min, se considera no-orilla segura.
Estos parmetros se pasan a openCV con la funcin _Canny_ (imagen, umbralMax, umbralMin)

```python
edges = cv2.Canny(img,180,250, apertureSize=3)
plt.imshow(edges,cmap = 'gray')
plt.show()
```
