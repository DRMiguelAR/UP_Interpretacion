__ _Pregunta:_ __ _¿En qué se diferencia la información contenida en una imagen en blanco y negro una a colores?_

La imagen se puede entender como una función bidimensional: $I(x,y)$, donde $I$ se relaciona con la intensidad de la imágen en un punto coordenado $(x,y)$. <br>
Si representamos esta imagen como una función bidimensional obtenemos la siguiente gráfica:

```r
img<-readPNG("galBW3.png")
#m<-img@red
mat1<-apply(img,2,rev)
colores=grey(seq(0,1,length=256))
#image(t(mat1), axes=FALSE,col=colores)
mat2=mat1[nrow(mat1):1,]
persp3D(z = mat2,phi=60, theta=90, col=colores)
png(filename="graph2.png")
persp3D(z = mat2,phi=60, theta=90, col=colores)
dev.off()
```

Al verlas una al lado de la otra,
